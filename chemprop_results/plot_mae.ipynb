```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_mae = pd.read_csv('mae.csv')
df_mae
```

```python
props = [
    "LogD",
    "LogS",
    "Log_HLM_CLint",
    "Log_MLM_CLint",
    "Log_Caco_Papp_AB",
    "Log_Caco_ER",
    "Log_Mouse_PPB",
    "Log_Mouse_BPB",
    "Log_Mouse_MPB",
]
```

```python
for prop in props:
    plt.figure(figsize=(10, 6))
    plt.barh(df_mae['model'], df_mae[f'{prop}_mae'], color='skyblue')
    plt.title(f'Mean Absolute Error for {prop}')
    plt.ylabel('Model')
    plt.xlabel('MAE')
    plt.tight_layout()
    plt.show()
```
